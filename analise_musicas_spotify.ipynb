```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("spotify_songs_dataset.csv")


df.dtypes
df.head(10)
```

```python
import matplotlib.pyplot as plt
#Relacionando Popularidade com conteudo explicito

tab_artista = df[["explicit_content","popularity"]]
sns.boxplot(x="explicit_content", y="popularity", data=tab_artista)

plt.title("Correlação de Conteúdo Explicito x Popularidade")
plt.xlabel("Popularidade")
plt.ylabel("Conteudo Explicito")
plt.show()
```

```python
#Genero mais Popular (bar chart)
tab_genre = df[["genre","stream","popularity" ]].dropna()

genre_popularity = tab_genre.groupby("genre")["popularity"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.bar(genre_popularity.index, genre_popularity.values)
plt.title("Genero mais Popular")
plt.xlabel("Genero")
plt.ylabel("Popularidade")
plt.show()
```

```python
#Genero mais Popular (line chart)
tab_genre = df[["genre","stream","popularity" ]].dropna()

genre_popularity = tab_genre.groupby("genre")["popularity"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.plot(genre_popularity.index, genre_popularity.values)
plt.title("Genero mais Popular")
plt.xlabel("Genero")
plt.ylabel("Popularidade")
plt.show()
```

```python
#Genero mais Popular (Scatter chart)
tab_genre = df[["genre","stream","popularity" ]].dropna()

genre_popularity = tab_genre.groupby("genre")["popularity"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.scatter(genre_popularity.index, genre_popularity.values)
plt.title("Genero mais Popular")
plt.xlabel("Genero")
plt.ylabel("Popularidade")
plt.show()
```

```python
tab_genre_count = df["genre"].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.bar(tab_genre_count.index, tab_genre_count.values)
plt.title("Gêneros com Mais Musicas")
plt.xlabel("Gênero")
plt.ylabel("Número de Músicas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
tab_scatter = df[["duration","popularity", "genre"]]

sns.scatterplot( x="duration", y="popularity", hue="genre", data=tab_scatter)
plt.xlabel("Duração")
plt.ylabel("Popularidade")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['popularity'], bins=35, kde=True)
plt.title('Distribuição da Popularidade das Músicas')
plt.xlabel('Popularidade')
plt.ylabel('Frequência')
plt.show()
```

```python
tab_soma = df[["artist","stream"]]

sum = tab_soma.groupby("artist")["stream"].sum().reset_index(0)
sum = sum.sort_values(by="stream", ascending=False)

sum["Percentual de Relevancia"] = sum["stream"]/sum["stream"].sum()*100
sum["Percentual de Relevancia"] = sum["Percentual de Relevancia"].round(4) 

sum.head(20).reset_index()
plt.figure(figsize=(12,6))
sns.barplot(data=sum.head(10), x="Percentual de Relevancia", y="artist", palette="viridis")
plt.title("Top 10 Artistas por Percentual de Relevância nos Streams")
plt.xlabel("Percentual de Relevância (%)")
plt.ylabel("Artista")
plt.show()
```
